### Статистика по языкам программирования на Июнь 2019

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 16,10
```

```python
a = pd.read_csv('C:/Users/Kanris/Desktop/salpr/2019_june_final.csv')
a.head(10)
```

```python
a = a.fillna('Unkn')
```

```python
b = a['Язык.программирования'].value_counts().sort_values(ascending=True).reset_index()
```

```python
plt.barh(b['index'],b['Язык.программирования'])
plt.ylabel('Язык программировани')
plt.xlabel('Количество респондентов')
plt.title('Распределение респондентов по языкам программирования',fontsize = 20)
plt.grid()
```

#### В построении данного графика участвуют 2 переменные: название языка программирования и количество опрошеных. Цель графика определить распределение респондентов (количество) по языкам программирования и выявить самые популярные из языков. Необходимо обратить внимание на тот факт, что более 5 500 респондентов не указали язык программирования, с которым они работают

```python
rasp = b[b['index'] != 'Unkn'].sort_values(by = 'Язык.программирования', ascending = False).head(10)
```

```python
sal = a.loc[a['Язык.программирования'].isin(rasp['index'])]
```

```python
sns.boxplot(data = sal, y = sal['Зарплата.в.месяц'], x = sal['Язык.программирования'])
plt.title('Распределение зарплат по ТОП 10 языков программирования',fontsize = 20)
```

#### Данный график отображает распределение зарплат по ТОП 10 языкам( поп популярности у респондентов).  По представленным данным можно сделать вывод, что разброс зарплат , находящихся выше медианного значения намного выше, чем у первой половины опрошенных. Зарплата распределяется неравномерно.   Необходим дополнительный анализ данных показателей( возможно такой размах  3-й и особенно 4-й квартили обусловлен нелинейным ростом зарплат в зависимости от опыта или иных факторов). Так же C# Java и JavaScript имеет большое количество выбросов, что так же требует дополнительного анализа( возможно была работа над специфическими проектами либо выбросы обсуловлены дополнительными компетенциями специалистов) 

```python
sns.heatmap(a.corr(),xticklabels=a.corr().columns, yticklabels=a.corr().columns, cmap='RdYlGn', center=0, annot=True)
plt.title('Корреляция между признаками опрошенных', fontsize = 20)
```

#### На данном графике представлена корреляция между всеми возможными парами переменных опроса. Можем сделать вывод , что наблюдается средняя по силе связь между опытом (exp) и зарплатой(slary) респондентов (коэф = 0.66) . Так же заметна достаточно слабая обратная связь между меткой "студент" и зарплатой( можем выдвинуть теорию о том, что студенты в целом по статистике получают меньшую ЗП по сравнению с другими специалистами, но некоторые могут зарабатывать на одном с ними уровне).

```python
sns.jointplot(x = 'exp', y = 'salary', data = a, kind = 'reg', height=10)
plt.title("Распределение зарплат в зависимости от опыта", fontsize=20)
```

#### На графике выше представлено распределение зарплат в зависимости от опыта специалиста. Согласно линии регрессии с опытом растет и зарплата сотрудника. Так же мы можем видеть, что наибольшее количество сотрудников имеет зарплату в районе 2.000, после чего их количество резко идет на спад с одновременным увеличением зарплаты. Согласно распределению , чем выше опыт сотрудника, тем меньше всего таких сотрудников. Рост количества сотрудников наблюдается до 3-х лет, после чего идет на спад. Однако, количество сотрудников со стажем 10 и более лет значительно выше предыдущих показателей и находится примерно на уровне 4-х лет. В целом, мы можем говорить о том, что зарплата зависит от опыта сотрудника.

```python
obl = a['Предметная.область'].value_counts().sort_values(ascending=False).reset_index(name='counts').head(10)
```

```python
import random
n = obl['index'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)
plt.bar(obl['index'],obl['counts'],color = c,  width=.5)
for i, val in enumerate(obl['counts'].values):
    plt.text(i, val, float(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':12})
plt.gca().set_xticklabels(obl['index'], rotation=60, horizontalalignment= 'right')
plt.title("ТОП 15 предметных областей", fontsize=20)
plt.ylabel('Количество респондентов')
```

#### Согласно данным лидирующую позицию среди предметных областей занимает Веб-разработка. Так же следует заметить, что в 5 из 15 областей встречается сочетание Веб-разработка + что-либо, это говорит о высокой популярности разработки в данном направлении на текущий момент.
